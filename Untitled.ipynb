```python
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle
from visual_genome import api as vg
from PIL import Image as PIL_Image
import requests
from StringIO import StringIO
```

```python
ids = vg.get_image_ids_in_range(0,1)
image_id = 61511
print "We got an image with id: %d" % image_id
%matplotlib inline
```

```python
image = vg.get_image_data(id=image_id)
print "The url of the image is: %s" % image.url
```

```python
regions = vg.get_region_descriptions_of_image(id=image_id)
print vg.get_region_graph_of_region().objects
print "The second region descriptions is: %s" % regions[1].phrase
print "It is located in a bounding box specified by x:%d, y:%d, width:%d, height:%d" % (regions[1].x, regions[1].y, regions[1].width, regions[1].height)
```

```python
fig = plt.gcf()
fig.set_size_inches(18.5, 10.5)
def visualize_regions(image, regions):
    response = requests.get(image.url)
    img = PIL_Image.open(StringIO(response.content))
    plt.imshow(img)
    ax = plt.gca()
    for region in regions:
        ax.add_patch(Rectangle((region.x, region.y),
                               region.width,
                               region.height,
                               fill=False,
                               edgecolor='red',
                               linewidth=3))
        ax.text(region.x, region.y, region.phrase, style='italic', bbox={'facecolor':'white', 'alpha':0.7, 'pad':10})
    fig = plt.gcf()
    plt.tick_params(labelbottom='off', labelleft='off')
    plt.show()
visualize_regions(image, regions[:8])
```
